```python
import matplotlib.pyplot as plt
```

```python
def plot_loss_ppl(filepath):
    val_losses, val_ppls = [], []
    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if line.startswith('| end of epoch'):
                segments = line.split('|')
                val_losses.append(float(segments[3].split()[-1]))
                val_ppls.append(float(segments[4].split()[-1]))
    assert len(val_losses) == len(val_ppls)
    
    # plot
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4), dpi=160)
    # loss
    axs[0].plot(val_losses, color='xkcd:magenta')
    axs[0].set_title('val loss as training progresses')
    axs[0].set_xlabel('epoch')
    axs[0].set_ylabel('loss')
    # perplexity
    axs[1].plot(val_ppls, color='xkcd:burnt orange')
    axs[1].set_title('val perplexity as training progresses')
    axs[1].set_xlabel('epoch')
    axs[1].set_ylabel('perplexity')
    plt.show()
```

## LSTM Model Embedding Size 100, Hidden Size 100, Dropout 0.2

```python
plot_loss_ppl('logs/lstm_emsize100_dropout0.2_tied.txt')
```

## LSTM Model Embedding Size 200, Hidden Size 200, Dropout 0.2

```python
plot_loss_ppl('logs/lstm_emsize200_dropout0.2_tied.txt')
```
