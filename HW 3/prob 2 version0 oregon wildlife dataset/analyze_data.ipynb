```
import os
import sys
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
import matplotlib as mpl
```

```
# change default text colour
COLOR = 'xkcd:bright sea green'
mpl.rcParams['text.color'] = COLOR
mpl.rcParams['axes.labelcolor'] = COLOR
mpl.rcParams['xtick.color'] = COLOR
mpl.rcParams['ytick.color'] = COLOR
```

```
# data directory (extracted path)
data_dir = os.path.join("gdrive/My Drive/Colab Notebooks/oregon wildlife/data", 
                        "oregon_wildlife", "oregon_wildlife")
# print folders in data directory
print(os.listdir(data_dir))
```

```
animals = os.listdir(data_dir)
numbers = list(map(lambda s: len(os.listdir(os.path.join(data_dir, s))), animals))
```

```
sum(numbers)
```

```
# plot
plt.figure(figsize=(4, 2), dpi=160)
plt.bar(range(1, len(animals) + 1), numbers, color='xkcd:warm purple')
plt.xticks(range(1, len(animals) + 1), animals, rotation=90)
plt.xlabel('wild animal')
plt.ylabel('number of samples')
plt.title('distribution of samples')
plt.show()
```

```
# select wild animals having at least one sample
selected_animals_dict = {}
for animal, number in zip(animals, numbers):
  if number > 0:
    selected_animals_dict[animal] = number
```

```
# plot
plt.figure(figsize=(4, 2), dpi=160)
plt.bar(range(1, len(selected_animals_dict) + 1), selected_animals_dict.values(), color='xkcd:warm purple')
plt.xticks(range(1, len(selected_animals_dict) + 1), selected_animals_dict.keys(), rotation=90)
plt.xlabel('wild animal')
plt.ylabel('number of samples')
plt.title('distribution of samples')
plt.show()
```
