# Data Augmentation

```python
import PIL
from PIL import Image
from torchvision import transforms
import matplotlib.pyplot as plt
```

```python
# define transformation for data augmentation
transform = transforms.RandomChoice(
    [
        transforms.ColorJitter(
            brightness=1,
            contrast=1,
            saturation=1,
            hue=0.5
        ),
        transforms.RandomGrayscale(p=1.0),
        transforms.RandomHorizontalFlip(p=1.0),
        transforms.RandomRotation(30, resample=PIL.Image.BILINEAR),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1))
    ]
)
```

```python
# method to augment data
def gen_augimage(image_path):
    """
    resizes and generates augmentation image
    
    :param image_path: path to image
    :return: resized original image, resized augmentation image
    """
    image = Image.open(image_path)  # PIL image
    
    # resize to 128 x 128
    image_resized = transforms.Resize(size=(128, 128))(image)
    
    # create generated image
    image_generated = transform(image_resized)
    return image_resized, image_generated

```

```python
# plot
def plot_images(images, title):
    fig, axs = plt.subplots(nrows=2, ncols=4, figsize=(6, 3), dpi=160, sharey=True)
    axs[0, 0].imshow(images[0][0])
    axs[0, 0].set_ylabel('original')
    axs[0, 1].imshow(images[1][0])
    axs[0, 2].imshow(images[2][0])
    axs[0, 3].imshow(images[3][0])
    axs[1, 0].imshow(images[0][1])
    axs[1, 0].set_ylabel('augmented')
    axs[1, 1].imshow(images[1][1])
    axs[1, 2].imshow(images[2][1])
    axs[1, 3].imshow(images[3][1])
    
    plt.suptitle(title)
    plt.show()
```

```python
# group A images
image_paths = ['data/gtsrb-german-traffic-sign/Train/0/00000_00000_00022.png',
               'data/gtsrb-german-traffic-sign/Train/1/00001_00000_00022.png',
               'data/gtsrb-german-traffic-sign/Train/2/00002_00000_00021.png',
               'data/gtsrb-german-traffic-sign/Train/3/00003_00000_00024.png']

images = []
for image_path in image_paths:
    images.append(gen_augimage(image_path))
    
plot_images(images, 'group A')
```
